The purpose of this project is to analyse wildfires in California between 2013 and 2020

# 2 - Changing datatypes
As can be seen from .info(), datatype related to datetime are not assigned, so we need to change the datatype to the correct Dtype:

```python
df['Extinguished'] = df['Extinguished'].astype('datetime64[ns]')
df['Started'] = df['Started'].astype('datetime64[ns]')
df['Updated'] = df['Updated'].astype('datetime64[ns]')
```

```python
df.info()
```

# 3 - Exploring the data
Now that the datatypes are correctly assigned for datetime, it is time to explore the data.

```python
# Checking the counties with more fires:
df['Counties'].value_counts()
```

```python
# Are there any active fires?
df['Active'].value_counts()
```

```python
# Let's assign a new column to the data frame where it tells how many days have the fire been active
df['days_active'] = ((df['Started'] - df['Extinguished'])*(-1)).dt.days
```

```python
df[df['days_active']>1].head(50)
```

```python
# Which state had the most area burned?
df[['AcresBurned','Counties']].sort_values(by='AcresBurned',ascending=False)
```

```python
# Acres don't make sense at all so we need to convert to hectares:
df['AcresBurned'] = round(df['AcresBurned']/2.471, 1)
df.rename({'AcresBurned': 'HectaresBurned'}, axis=1, inplace=True)
```

```python
df.head()
```

```python
# Checking that the calculations are correct:
df.loc[df['CanonicalUrl'] =='/incidents/2018/12/31/bridge-fire/']
```

# 4 - Cleaning the data
Since we are not going to need all the features, some of the columns and rows can be dropped.

```python
df.head()
```

```python
# Since all the actives fires = False, this column can be dropped.
len(df[df['Active'].notnull()])
```

```python
# Most of the control statement are null, hence this column can also be dropped.
len(df[df['ControlStatement'].notnull()])
```

```python
# So, featured can also be dropped.
print(f"Featured = True represent {round(len(df[df['Featured']==True])/len(df[df['Featured']==False])*100,0)}% of the data")
```

```python
# Final also represents 100% of the data but is not necessary, so:
round(len(df[df['Final']==False])/len(df[df['Final']==True]),0)*100
```

```python
# What about fuel type?
print(f"Fuel types represents {len(df['FuelType'])/len(df['FuelType']*100)}% of the column, 99% are Null")
```

# 5 - Answering questions
Now that the dataframe is cleaned and organized, it is time to start answering some questions with plots and graphs

### 5.1 - At what time do usually wild fires occurr?

```python
# Let's first have a look once more at the dataframe:
cal_fires.head()
```

```python
# In order to find the mean of the time, we have to deduct the normalized datetime and then calculate mean of timedelta series
mean_time = cal_fires['Started'] - cal_fires['Started'].dt.normalize()
res = str(mean_time.mean())[-18:-13]
print(f'Wild fires in California usually start at {res} hours')
```

### 5.2 - How frequent and destructive wild fires are?
To answer this question, we need to find how many fire occurr in a year, and the hectares affected.

```python
# total fires
cal_fires['Started'].count()
```

```python
cal_fires['YearStarted'] = cal_fires['Started'].dt.year
cal_fires['MonthStarted'] = cal_fires['Started'].dt.month
```

```python
cal_fires['YearStarted'].value_counts().sort_index().plot(kind='bar', figsize=(12,7))
plt.xlabel('Year')
plt.ylabel('Number of Fire Incidents')
plt.title('Frequency of Fire Incidents per Year Between 2013-2019');
```

```python
cal_fires['MonthStarted'].value_counts().sort_index().plot(kind='bar', figsize=(12,7))
plt.xlabel('Year')
plt.ylabel('Number of Fire Incidents')
plt.title('Frequency of Fire Incidents per Year Between 2013-2019');
```

```python
cal_fires[cal_fires['YearStarted'] == 1969]
cal_fires = cal_fires.drop(cal_fires[cal_fires.YearStarted == 1969].index)
```

```python
cal_fires['YearStarted'].value_counts().sort_index().plot(kind='bar', figsize=(12,7))
plt.xlabel('Year')
plt.ylabel('Number of Fire Incidents')
plt.title('Frequency of Fire Incidents per Year Between 2013-2019');
```

```python
cal_fires.plot(x="YearStarted", y="HectaresBurned").plot(kind='bar', figsize=(12,7))
plt.show()
```
